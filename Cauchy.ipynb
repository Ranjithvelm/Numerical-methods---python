```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = [1.992, 7.2*10**(4), -5.1*10**(-28)]
B = [0.038, -3.4*10**(-7), 1.4*10**(-13 )]
Lam = np.linspace(700,1700,2000)
```

```python
n = A[0] + A[1]/Lam**2 + A[2]/Lam**4
k = B[0] + B[1]/Lam**2 + B[2]/Lam**4
print(n)
print(k)
```

```python
plt.plot(Lam, n)
```

```python
plt.plot(Lam, k)
```
